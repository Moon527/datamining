```python
import pandas as pd
df = pd.read_csv('/content/labeled_data.csv')
df.head(5)
```

```python
#데이터 전처리
df = df[df['PART_NAME'] != 'SP2 CVR ROOF RACK CTR, RH']
df = df[df['PART_NAME'] != "JX1 W/S SIDE MLD'G RH"]
df = df[df['EQUIP_NAME'] == '650톤-우진2호기']
df = df.drop(columns=['Reason', '_id', 'TimeStamp', 'PART_FACT_PLAN_DATE', 'PART_FACT_SERIAL', 'EQUIP_NAME', 'EQUIP_CD'])
#모든행이 다 0이면 삭제
df = df.loc[:, (df != 0).any(axis=0)]
df.head()
```

```python
cn7 = df[df['PART_NAME'].str.startswith('CN7')]
rg3 = df[df['PART_NAME'].str.startswith('RG3')]
```

```python
cn7['PassOrFail'] = cn7['PassOrFail'].replace('Y',0).replace('N',1)
```

```python
# prompt: cn7 모든 변수 박스플롯 칸마다 나눠서 박스플롯 세로로

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'cn7' DataFrame is already defined as in the previous code

# Create subplots for each column
num_cols = len(cn7.columns)
fig, axes = plt.subplots(num_cols, 1, figsize=(10, 5 * num_cols))

# Iterate over columns and create boxplots
for i, col in enumerate(cn7.columns):
  sns.boxplot(ax=axes[i], y=cn7[col])
  axes[i].set_title(f"Boxplot of {col}")
  axes[i].set_ylabel(col)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib_inline
import seaborn as sns
cn7_ = cn7.drop(columns=['PART_NAME'])
plt.subplots(figsize=(25,25))
sns.heatmap(data = cn7_.corr(), linewidths=0.1, annot=True, fmt ='.2f', cmap='Blues')
```

```python
n_cols = 4
n_rows = (len(cn7_.columns) + n_cols - 1) // n_cols
plt.figure(figsize=(5 * n_cols, 4 * n_rows))
for idx, col in enumerate(cn7_.columns):
    plt.subplot(n_rows, n_cols, idx + 1)
    plt.hist(cn7_[col].dropna(), bins=30, color='skyblue', edgecolor='black')
    plt.title(col)
    plt.tight_layout()

plt.show()
```

```python
# 1. 데이터 이산화
df_discrete = cn7[['Injection_Time','Cycle_Time','Max_Injection_Pressure']]
df_discrete = df_discrete.apply(lambda x: pd.cut(x, bins=3, labels=['Low','Mid','High']))

# 2. 연관 규칙 추출 (Apriori)
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(pd.get_dummies(df_discrete), min_support=0.05, use_colnames=True)

# 3. 규칙 해석
print(frequent_itemsets.sort_values('support', ascending=False).head(5))
```

```python
from sklearn.tree import DecisionTreeClassifier

# 1. 데이터 준비
X = cn7[['Injection_Time','Cycle_Time','Max_Injection_Pressure']]
y = cn7['PassOrFail']

# 2. 모델 학습 (깊이 3으로 제한)
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(X, y)

# 3. 시각화
from sklearn.tree import plot_tree
plt.figure(figsize=(15,8))
plot_tree(dt, feature_names=X.columns, filled=True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1. 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(cn7[['Injection_Time','Cycle_Time']])

# 2. 클러스터링 실행
kmeans = KMeans(n_clusters=3)
cn7['Cluster'] = kmeans.fit_predict(X_scaled)

# 3. 군집 특성 분석
cluster_stats = cn7.groupby('Cluster')['PassOrFail'].mean()
print(cluster_stats)
```

```python
from sklearn.linear_model import LinearRegression

# 1. 변수 선택
X = cn7[['Injection_Time','Max_Injection_Pressure']]
y = cn7['Cycle_Time']

# 2. 모델 학습
model = LinearRegression()
model.fit(X, y)

# 3. 계수 해석
coeff = pd.Series(model.coef_, index=X.columns)
print(coeff.sort_values(ascending=False))
```
